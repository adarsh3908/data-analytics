```python
import pandas as pd

listings = pd.read_csv(
    r"C:\Users\ASUS\Downloads\Airbnb+Data\Airbnb Data\Listings.csv",
    encoding='latin1', 
    low_memory=False,parse_dates=["host_since"] )
```

```python
listings.head()
```

```python
listings.info()
```

```python
listings["host_since"] = pd.to_datetime(listings["host_since"])
```

```python
paris_listings = (
      listings
      .query("city == 'Paris'")
        .loc[:,["host_since","neighbourhood","city","accommodates","price"]])
paris_listings.info()
```

```python
paris_listings.isna().sum()
```

```python
paris_listings.describe()
```

```python
paris_listings.query("accommodates == 0").count()
```

```python
paris_listings_neighborhood =(
paris_listings.groupby("neighbourhood").agg({"price":"mean"}).sort_values("price"))
paris_listings_neighborhood.tail()
```

```python
paris_listings_accommodates = (
paris_listings.query("neighbourhood == 'Elysee'").groupby("accommodates")
.agg({"price":"mean"}).sort_values("price"))

paris_listings_accommodates.head()
```

```python
paris_listings_over_time = (
    paris_listings.set_index("host_since").resample("Y")
    .agg({
        "price":"mean","neighbourhood":"count"
    })
)
paris_listings_over_time.head()
```

```python
import seaborn as sns
(paris_listings_neighborhood.plot
.barh(title = "Average Listing Price by Paris Neighbourhood",
      xlabel = "Price Per Night (Euros)",
      ylabel = "Neighbourhood",
      legend = None
))
sns.despine()
```

```python
(paris_listings_accommodates.plot
.barh(title = "Average Listing Price by Accomodation Number",
      xlabel = "Price Per Night (Euros)",
      ylabel = "Neighbourhood",
      legend = None
))
sns.despine()
```

```python
paris_listings_over_time["neighbourhood"].plot(
ylabel="New Hosts",
title ="New AirBnb Hosts in Paris Over Time")
sns.despine()
```

```python
paris_listings_over_time["price"].plot(
ylabel="Average Price (Euros)",
title ="Average AirBnb Price in Paris Over Time")
sns.despine()
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.plot(paris_listings_over_time.index,
        paris_listings_over_time["neighbourhood"],
        label= "New Hosts",
        c="pink"
)
ax.set_ylabel("New Hosts")
ax2 = ax.twinx()
ax2.plot(paris_listings_over_time.index,
        paris_listings_over_time["price"],
        label= "Average Price",
        
)
ax.set_ylabel("Average Price")
ax.set_title("2015 Regulations Lead to Fewer New Hosts, Higher Prices")
```
